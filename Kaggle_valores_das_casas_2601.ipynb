```python
# Criando variável com a quantidade de linhas
qtd_linhas = len(bd_ajuste)
print(qtd_linhas)
```

```python
# Criando lista com as colunas
colunas = []
colunas_removidas = []
for column in bd_ajuste:
    colunas.append(column)
print(colunas)
print(colunas_removidas)
```

```python
#Estabelecendo ponto de corte de colunas vazias
porcentagem = cortex
corte = round((porcentagem/100) * qtd_linhas)
print('Numero de corte: ',corte)
```

```python
from pandas.core.indexes.datetimes import bdate_range
#Remoção de colunas com ausencia de dados
for col in colunas:   
    prov = bd_ajuste[col]
    contagem_valores = prov.count() 
    if (contagem_valores < corte):
        print(col, contagem_valores, 'Removida do banco de dados')
        bd_ajuste.drop([col], axis="columns", inplace=True)
        colunas_removidas.append(col)
        colunas.remove(col)
        print('Coluna removida: ', col)
    else:
        print(col, contagem_valores)
print(colunas)
```

```python
# Preenchendo valores das colunas numéricas (ausência de dados) - train
for col in colunas:   # PRECISO ENTENDER ESSE FOR pois depois deste comando 'bd_ajuste.info(81)' ainda continuaram valores Non-Null Count ai depois preenche com valores "0" onde ainda faltou
    prov = bd_ajuste[col] #Cria o dataframe prov com a coluna do laço for
    if (is_string_dtype(bd_ajuste[col]) == True): #Testa se a coluna é do do tipo string
        vazio = 'vazio'+col #Cria a variavel vazio = 'Vazio' + nome da coluna (para evitar a criação de colunas do mesmo nome)
        prov.fillna(vazio, inplace= True) #Substitui o campo vazio pela variavel vazio
        bd_ajuste.drop([col], axis="columns", inplace=True) #Apaga a coluna do laço for do banco de dados
        bd_ajuste[col] = prov #Adiciona a coluna do dataframe prov no bd_ajuste (maneira que fiz para substituir a coluna com as dados vazios)
        print(col, 'String')  #Imprime o nome da coluna e o tipo string
    elif (is_numeric_dtype(train[col]) == True):  #Testa se a coluna é do do tipo numérico
        mediana = prov.median() #calcula a mediana do dataframe prov (Com apenas a coluna em estudo)
        prov.fillna(mediana, inplace= True) #Substitui o campo vazio pela mediana
        contagem_valores = prov.count() #Cria uma variavel com a contagem dos campos preenchidos
        prov = prov.astype(np.int64) #Transforma a coluna em inteiros
        bd_ajuste.drop([col], axis="columns", inplace=True)  #Apaga a coluna do laço for do banco de dados
        bd_ajuste[col] = prov    #Adiciona a coluna do dataframe prov no bd_ajuste             
        print(col,mediana, contagem_valores)  #Imprime a coluna, mediana e a quantidade de valores
    else:
        print(col, 'Outros') #Comando para mostrar se a coluna não se enquadra no if ou else (caso de ser variável boleana)
```

7 - Testar melhor algoritmo

```python
#Bando de dados dos resultados
y_train = trainZ['SalePrice']
y_train.head(5)
```

```python
#Banco de dados das variáveis
X_train = trainZ.drop('SalePrice', axis = 1)
X_train.info()
```

```python
#Balanceando valores
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_train_scaled 
```

```python
#Treinamento
from sklearn.model_selection import cross_val_score
cv = lotes
scoring = 'neg_mean_squared_error'
import multiprocessing
n_jobs = multiprocessing.cpu_count() - 1
n_jobs
```

```python
#Teste algoritmo Regressão linear
from sklearn.linear_model import Ridge
model = Ridge()
scores = cross_val_score(model, X_train_scaled, y_train, cv = cv,
                         scoring = scoring, n_jobs = n_jobs)
linear = np.sqrt(-scores.mean())
linear
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
scores = cross_val_score(model, X_train_scaled, y_train, cv = cv,
                         scoring = scoring, n_jobs = n_jobs)
tree = np.sqrt(-scores.mean())
tree
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
scores = cross_val_score(model, X_train_scaled, y_train, cv = cv,
                         scoring = scoring, n_jobs = n_jobs)
random = np.sqrt(-scores.mean())
random
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
scores = cross_val_score(model, X_train_scaled, y_train, cv = cv,
                         scoring = scoring, n_jobs = n_jobs)
gradient = np.sqrt(-scores.mean())
gradient
```

```python
#Deep Learning MLPRegressor
from sklearn.neural_network import MLPRegressor
model = MLPRegressor()
scores = cross_val_score(model, X_train_scaled, y_train, cv = cv,
                         scoring = scoring, n_jobs = n_jobs)
deep = np.sqrt(-scores.mean())
deep
```

```python
#Inspecionando o melhor modelo
from sklearn.model_selection import train_test_split
X_train1, X_test, y_train1, y_test = train_test_split(X_train_scaled, y_train,
                                                      test_size = amostrateste, random_state = 16)
```

```python
#Treinamento do modelo
model = GradientBoostingRegressor()
model.fit(X_train1, y_train1)
```
